```python
import re
# re- regular expression
#the given text is for example 
#wherever you encounter space split from the same
#/s denote space 
text ="Hello, world. This, is a test."
result = re.split(r'(/s)',text)

print(result)
```

```python
# wherever you encounter , . /s space and create separate for the same
result = re.split(r'([,.]|\s)',text)

print(result)
```

```python
#space token are also added remove them
#for this example removing space is good for memory and low computation
result = [item for item in result if item.strip()]

print(result)
```

```python
class SimpleTokenizerV1:
    def __init__(self, vocab):
        self.str_to_int = vocab         #unique token :  unique id
        self.int_to_str = {i:s for s,i in vocab.items()}  #unique id  :  unique token

    def encode(self,text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)',text) #special char found split the words

        preprocessed =[
            item.strip() for item in preprocessed if item.strip()  #remove whitespace token

        ]

        ids = [self.str_to_int[s] for s in preprocessed] #assign unique id to each token
        return ids
    
    def decode(self,ids): #
        text=" ".join([self.int_to_str[i] for i in ids])  #unique id  :  unique token and separate each token by whitespace

        text = re.sub(r'\s+([,.?!"()\'])',r'\1',text) #word space specialCharacter remove space from this
        return text
```

```python
tokenizer = SimpleTokenizerV1(vocab)

text=""""It's the last he painted, you know,"
      Mrs.Gisburn said with pardonable pride."""       #convert token into id

ids = tokenizer.encode(text)

print(ids)
```

```python
tokenizer.decode(ids)
```

```python
class SimpleTokenizerV2:
    def __init__(self,vocab):
        self.str_to_int = vocab
        self.int_to_str = {i:s for s,i in vocab.items()}

    #it will go through query and for query word token not available assign with <|unk|>
    def encode(self,text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)',text)
        preprocessed = [item.strip() for item in preprocessed if item.strip()]

        preprocessed =[
            item if item in self.str_to_int
            else "<|unk|>" for item in preprocessed
        ]

        ids =[self.str_to_int[s] for s in preprocessed]
        return ids
    def decode(self,ids):
        text=" ".join([self.int_to_str[i] for i in ids])

        text = re.sub(r'\s+([,.:;?!"()\'])',r'\1',text)
        return text
```

```python
tokenizer = SimpleTokenizerV2(vocab)

#reading of text1 completed <|endoftext|> and text2 started
text1 = "Hello, do you like tea?"
text2 = "In the sunlight terraces of the palace."

text =" <|endoftext|> ".join((text1,text2))

print(text)
```

```python
tokenizer.encode(text)
```

```python
id = tokenizer.encode(text)
print(id)
```

```python
tokenizer.decode(id)
```
